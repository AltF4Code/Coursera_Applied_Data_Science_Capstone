```python
#DataFrame Manipulation
import numpy as np
import pandas as pd

#Web Scrapping
import requests
from bs4 import BeautifulSoup
import xml
```

```python
page_req = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(page_req.content, 'html.parser')
```

```python
table_post = soup.find('table')
fields = table_post.find_all('td')

postcode = []
borough = []
neighbourhood = []

for i in range(0, len(fields), 3):
    postcode.append(fields[i].text.strip())
    borough.append(fields[i+1].text.strip())
    neighbourhood.append(fields[i+2].text.strip())
        
df_postal_codes = pd.DataFrame(data=[postcode, borough, neighbourhood]).transpose()
df_postal_codes.columns = ['Postcode', 'Borough', 'Neighbourhood']
df_postal_codes['Borough'].replace('Not assigned', np.nan, inplace=True)
df_postal_codes.dropna(subset=['Borough'], inplace=True)
```

```python
df_postal_codes_new = df_postal_codes.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df_postal_codes_new.columns = ['Postcode', 'Borough', 'Neighbourhood']
df_postal_codes_new['Neighbourhood'].replace('Not assigned', "Queen's Park", inplace=True)
```

```python
from geopy.geocoders import Nominatim
import folium
```

```python
df_geospat = pd.read_csv('http://cocl.us/Geospatial_data')
df_geospat.columns = ['Postcode', 'Latitude', 'Longitude']
```

```python
df_position = pd.merge(df_postal_codes_new, df_geospat, on=['Postcode'], how='inner')
df_geo = df_position[['Borough', 'Neighbourhood', 'Postcode', 'Latitude', 'Longitude']].copy()
df_geo.head()
```

```python
geolocator = Nominatim()
loc = geolocator.geocode('Toronto, Canada')
lat = loc.latitude
long = loc.longitude
```

```python
map_toronto = folium.Map(location=[lat, long], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_geo['Latitude'], df_geo['Longitude'], df_geo['Borough'], df_geo['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=3,
        popup=label,
        color='green',
        fill=True,
        fill_color='#3199cc',
        fill_opacity=0.3,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
